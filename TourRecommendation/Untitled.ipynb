```python
import pymysql
def dbConnection():
    try:
        connection = pymysql.connect(host="localhost", user="root", password="root", database="tour_recommendation")
        return connection
    except Exception as e:
        print(e)
        print("Something went wrong in database Connection")

def dbClose():
    try:
        dbConnection().close()
    except:
        print("Something went wrong in Close DB Connection")

con = dbConnection()
cursor = con.cursor()

cursor.execute('SELECT * FROM place_db WHERE rating > 4.5 ORDER BY rating DESC')
row = cursor.fetchall()
print(row)
print("HHHHHHHHHHHHHH")
cursor.execute('SELECT * FROM place_db WHERE rating < 4.5 ORDER BY rating ASC')
row1 = cursor.fetchall()
print(row1)
```
